2.Problem Statement

Explain the difference between Absolute and Relative positioning.

 Absolute Positioning

When you give an element position: absolute;, it is taken out of the normal document flow.

Its position is calculated relative to the nearest positioned ancestor (an ancestor that has position: relative, absolute, or fixed).

If no such ancestor exists, it is positioned relative to the <html> (viewport).

Example:

```
<div style="position: relative; width: 200px; height: 200px; background: lightgray;">
  <div style="position: absolute; top: 20px; left: 30px; width: 50px; height: 50px; background: red;">
    Box
  </div>
</div>
```

Relative Positioning

When you give an element position: relative;, it is still part of the normal document flow (it keeps its original place).

But you can shift it relative to its original position using top, left, right, bottom.

Other elements still behave as if it wasn’t moved.

Example:

```
<div style="position: relative; top: 20px; left: 30px; width: 100px; height: 100px; background: blue;">
  Box
</div>
```

It only works on positioned elements (position: absolute, relative, fixed, or sticky) and flex items (elements that are direct children of display:flex elements).

A higher z-index means the element will be stacked in front of elements with lower z-index values.
